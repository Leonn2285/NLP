```python
df.info()
```

```python
df = df.dropna(how="any")
df.info()
```

```python
import underthesea
```

```python
# Tách từ cột title
df['title'] = df['title'].apply(lambda x: ' | '.join(underthesea.word_tokenize(x)))
# Tách từ cột content
df['content'] = df['content'].apply(lambda x: ' | '.join(underthesea.word_tokenize(x)))
```

```python
import re

def clean_text(text):
    # Loại bỏ các ký tự không phải chữ cái, số, khoảng trắng và dấu |
    text = re.sub(r'[^\w\s|]', ' ', text)

    # Loại bỏ số
    text = re.sub(r'\d+', ' ', text)

    # Loại bỏ khoảng trắng thừa
    text = re.sub(r'\s+', ' ', text).strip()

    # Loại bỏ các dấu | thừa
    text = re.sub(r'\|+', '|', text)
    text = re.sub(r'^\||\|$', '', text)

    return text

df['title'] = df['title'].apply(clean_text)
df['content'] = df['content'].apply(clean_text)
```

```python
combined_df_by_category = df.groupby('topic').agg({
    'id': lambda x: ' | '.join(x),
    'title': lambda x: ' | '.join(x),
    'content': lambda x: ' | '.join(x)
}).reset_index()

display(combined_df_by_category)
```

### Câu 1 - Tạo một bảng dữ liệu trong đó, mỗi từ là một feature, và giá trị của nó là số lần xuất hiện của nó. Có thể cộng cả content+title lại. Tiếp đến feed bảng dữ liệu này vào 3 mô hình học máy của scikit (Logistic, SVM, NaiveBayes)

```python
# Kết hợp title và content để tạo text đầy đủ
df['text'] = df['content'] + ' | ' + df['title']

combined_df = df[['topic', 'text']]
print(f"\nSố lượng samples: {len(combined_df)}")
display(combined_df)
```

```python
mega_docs = ""
for i in range(len(combined_df)):
  mega_docs += combined_df["text"].iloc[i]

vocab = mega_docs.split(" | ");
vocab_set = set(vocab)

# Loại bỏ '' và loại bỏ ký tự '|' trong từng phần tử
vocab_set = {item.replace('|', '').strip() for item in vocab_set if item.strip() != ''}

print(vocab_set)
```

```python
vocab_list = list(vocab_set)
word_features = pd.DataFrame(np.zeros([len(combined_df), len(vocab_set)]), columns=vocab_list)
word_features['topic'] = combined_df['topic']

print(word_features.shape)
```

```python
# Tối ưu hóa: dùng vectorization thay vì 2 vòng lặp lồng nhau
from collections import Counter

for i in range(len(combined_df)):
    text_lower = combined_df["text"].iloc[i].lower()
    # Đếm số lần xuất hiện của tất cả các từ trong text
    word_count = Counter([w for w in vocab_list if w.lower() in text_lower])

    # Cập nhật toàn bộ row một lần thay vì từng cell
    for word in word_count:
        word_features.at[i, word] = text_lower.count(word.lower())
```

```python
display(word_features)
```

```python
y = word_features['topic']
X = word_features.drop('topic', axis=1)

print(f"Shape of X (features): {X.shape}")
print(f"Shape of y (target): {y.shape}")
```

```python
# Kiểm tra NaN trước khi train
print("Số NaN trong X:", X.isna().sum().sum())
print("Số NaN trong y:", y.isna().sum())
print(f"\nShape trước khi xử lý: X={X.shape}, y={y.shape}")

# Loại bỏ các hàng có NaN
mask = ~(X.isna().any(axis=1) | y.isna())
X = X[mask]
y = y[mask]

print(f"Shape sau khi xử lý: X={X.shape}, y={y.shape}")
```

### Feed vào 3 mô hình

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
def TrainModel(model, _X_train, _X_test, _y_train, _y_test):
    model.fit(_X_train, _y_train)

    y_pred = model.predict(_X_test)

    acc = accuracy_score(_y_test, y_pred)
    prec = precision_score(_y_test, y_pred, average='macro', zero_division=0)
    rec = recall_score(_y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(_y_test, y_pred, average='macro', zero_division=0)
    print(f"Model {model} - Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

    return model
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB

logistic = TrainModel(LogisticRegression(solver='liblinear', random_state=42), X_train, X_test, y_train, y_test)
svm = TrainModel(SVC(kernel="sigmoid"), X_train, X_test, y_train, y_test)
nb = TrainModel(MultinomialNB(), X_train, X_test, y_train, y_test)
```
