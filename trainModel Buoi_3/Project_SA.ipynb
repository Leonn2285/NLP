### Câu 1 - Tạo một bảng dữ liệu trong đó, mỗi từ là một feature, và giá trị của nó là số lần xuất hiện của nó. Có thể cộng cả content+title lại. Tiếp đến feed bảng dữ liệu này vào 3 mô hình học máy của scikit (Logistic, SVM, NaiveBayes)

```python
import re

def clean_text(text):
    # Loại bỏ các ký tự không phải chữ cái, số, khoảng trắng và dấu |
    text = re.sub(r'[^\w\s|]', ' ', text)

    # Loại bỏ số
    text = re.sub(r'\d+', ' ', text)

    # # Loại bỏ khoảng trắng thừa
    # text = re.sub(r'\s+', ' ', text).strip()

    # Loại bỏ các dấu | thừa
    text = re.sub(r'\|+', '|', text)
    text = re.sub(r'^\||\|$', '', text)

    return text

combined_df['title_clean'] = combined_df['title'].apply(clean_text)
combined_df['content_clean'] = combined_df['content'].apply(clean_text)
```

```python
# Bước 1: Kết hợp title và content để tạo text đầy đủ
combined_df['full_text'] = combined_df['title_clean'] + ' ' + combined_df['content_clean']
print(f"\nSố lượng samples: {len(combined_df)}")
display(combined_df[['topic', 'full_text']].head())
```

```python
# Tách cột full_text thành list từ
combined_df['words'] = combined_df['full_text'].str.split('|')

# Tách dòng (explode) để mỗi từ thành một hàng
df_exploded = combined_df.explode('words')

# Xóa khoảng trắng dư
df_exploded['words'] = df_exploded['words'].str.strip()

# Đếm số lần từ xuất hiện trong từng topic
df_count = pd.crosstab(df_exploded['topic'], df_exploded['words'])

# Reset index
df_count = df_count.reset_index()

# Hiển thị kết quả
df_count.head()
```

### Feed vào 3 mô hình 

```python
y = df_count['topic']
X = df_count.drop('topic', axis=1)

print(f"Shape of X (features): {X.shape}")
print(f"Shape of y (target): {y.shape}")
```

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.1, random_state=42)
```

```python
def TrainModel(model, _X_train, _X_test, _y_train, _y_test):
    model.fit(_X_train, _y_train)

    y_pred = model.predict(_X_test)

    acc = accuracy_score(_y_test, y_pred)
    prec = precision_score(_y_test, y_pred, average='macro', zero_division=0)
    rec = recall_score(_y_test, y_pred, average='macro', zero_division=0)
    f1 = f1_score(_y_test, y_pred, average='macro', zero_division=0)
    print(f"Model {model} - Accuracy: {acc:.4f}, Precision: {prec:.4f}, Recall: {rec:.4f}, F1: {f1:.4f}")

    return model
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB

logistic = TrainModel(LogisticRegression(solver='liblinear', random_state=42), X_train, X_test, y_train, y_test)
svm = TrainModel(SVC(kernel="sigmoid"), X_train, X_test, y_train, y_test)
nb = TrainModel(MultinomialNB(), X_train, X_test, y_train, y_test)
```
