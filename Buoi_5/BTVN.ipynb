## 2/ Tách câu để tạo thành 1 list, mỗi phần tử là một dòng

```python
# Tách theo dấu chấm
lines = text.split('.')
# Xử lý từng dòng: thay thế xuống dòng bằng dấu cách và loại bỏ dòng trống
lines = [line.replace('\n', ' ').strip() for line in lines if line.strip()]
print(lines)
```

```python
import json

# Tạo cấu trúc JSON cho Label Studio
data = []
for line in lines:
    data.append({
        "text": line
    })

# Lưu vào file data.json
with open('data.json', 'w', encoding='utf-8') as f:
    json.dump(data, f, ensure_ascii=False, indent=2)
```
