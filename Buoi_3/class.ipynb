```python
def gender_features(word):
    return {'last_letter': word[-1]}
gender_features('Shrek')
```

```python
import nltk
from nltk.corpus import names 
labelled_names = ([(name, 'male') for name in names.words('male.txt')] +
                  [(name, 'female') for name in names.words('female.txt')])
import random
random.shuffle(labelled_names)
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labelled_names]
train_set, test_set = featuresets[500:], featuresets[:500]
classifier = nltk.NaiveBayesClassifier.train(train_set)
```

```python
print(nltk.classify.accuracy(classifier, test_set))
```

**sklearn**

```python
from sklearn.linear_model import LogisticRegression
```

```python
featuresets = [(gender_features(n), gender) for (n, gender) in labelled_names]
train_set, test_set = featuresets[500:], featuresets[:500]
nb_classifier = nltk.NaiveBayesClassifier.train(train_set)
tree_classifier = nltk.DecisionTreeClassifier.train(train_set)
logreg_classifier = nltk.classify.SklearnClassifier(LogisticRegression())
logreg_classifier.train(train_set)
```

```python
print(tree_classifier.classify(gender_features('Neo')))
print(tree_classifier.classify(gender_features('Trinity')))
print(logreg_classifier.classify(gender_features('Trinity')))
```

```python
print(nltk.classify.accuracy(tree_classifier, test_set))
print(nltk.classify.accuracy(nb_classifier, test_set))
print(nltk.classify.accuracy(logreg_classifier, test_set))
```
