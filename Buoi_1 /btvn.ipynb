**Lấy 1 đoạn văn bản trên wiki, sau đó thực thi các công việc sau**
- Tách từ với 2 cấp độ, từ đơn, từ ghép 2 chữ. (naive)
- Phương pháp 1 : Sử dụng 1 bộ từ điển tiếng việt mở để loại bỏ những từ không tồn tại trong từ điển
- Phương pháp 2 : Sử dụng 1 trong 2 thư viện pyvi hoặc underthesea để tách từ 
- Thống kê số lượng từ tách đúng bằng 2 phương pháp nói trên 

```python
s = "Đã đi qua những ngày tết cổ truyền, tôi lại bước chân lên tàu và đi đến một miền đất xa xôi mà tôi đã chọn để học tập, tôi đi xa bà, xa ông, xa bạn bè, đặc biết nhất là xa quê hương. Ôi! Hai tiếng quê hương! Nhớ quê! Tôi chỉ biết khóc tôi thấy đâu đây vị ngọt ngào của nước mắt, chính quê hương đã ban cho tôi giọt nước mắt ngọt ngào đó, Ngày mai, tôi sẽ đi xa nơi đây đến phương trời kia không phải là phương trời quen thuộc như mỗi lần tôi nằm dưới bãi cỏ và ngắm nhìn bầu trời xanh kia. Đi! Thật xa gặp những con người mới ở xứ lạ. Tôi sẽ cố gắng học thật tốt dưới mảnh đất xa lạ ấy, quê hương tôi nằm ở đây trong con tim tôi đây này."
```

```python
import re

words_single = re.findall(r"\w+", s.lower())
print(f"Số từ đơn: {len(words_single)}")
print(f"Các từ đơn: {words_single[:20]}...")  

# Tách từ ghép 2 chữ (bigram)
words_bigram = []
words_list = s.lower().split()
for i in range(len(words_list) - 1):
    # Loại bỏ dấu câu
    word1 = re.sub(r'[^\w]', '', words_list[i])
    word2 = re.sub(r'[^\w]', '', words_list[i+1])
    if word1 and word2:
        bigram = f"{word1}_{word2}"
        words_bigram.append(bigram)

print(f"\nSố từ ghép 2 chữ: {len(words_bigram)}")
print(f"Các từ ghép 2 chữ: {words_bigram[:20]}...")  
```

## Phương pháp 1: Sử dụng từ điển tiếng Việt

```python
# Tải từ điển tiếng Việt
from pyvi import ViTokenizer

# Đọc stopwords
with open("vietnamese_dict_online.txt", 'r', encoding='utf-8') as f:
    vietnamese_dict = set(line.strip().lower() for line in f if line.strip())

# Thêm từ ghép phổ biến
compound_words = ["quê hương", "bạn bè", "học tập", "cố gắng", "con người",
                  "bầu trời", "nước mắt", "ngọt ngào", "bãi cỏ", "xứ lạ",
                  "cổ truyền", "xa xôi", "đặc biệt"]
vietnamese_dict.update(compound_words)

print(f"Từ điển: {len(vietnamese_dict)} từ (stopwords + từ ghép)")
```

```python
# So sánh với từ điển

# Lọc từ đơn
valid_single = [w for w in words_single if w in vietnamese_dict]

# Lọc bigram (kiểm tra cụm từ)
valid_bigram = [bg for bg in words_bigram if bg.replace('_', ' ') in vietnamese_dict]

print(f"Từ đơn: {len(valid_single)}/{len(words_single)} hợp lệ")
print(f"Bigram: {len(valid_bigram)}/{len(words_bigram)} có nghĩa")
print(f"  Ví dụ bigram hợp lệ: {valid_bigram[:5] if valid_bigram else 'Không có'}")
```

## Phương pháp 2: Sử dụng thư viện pyvi hoặc underthesea

```python
# Cài đặt thư viện (nếu chưa có)
# !pip install pyvi
# !pip install underthesea

# Sử dụng pyvi
from pyvi import ViTokenizer

tokenized_pyvi = ViTokenizer.tokenize(s)
print("Kết quả tách từ bằng pyvi:")
print(tokenized_pyvi)
print()

# Tách thành list các từ
words_pyvi = tokenized_pyvi.split()
print(f"Số từ tách được bằng pyvi: {len(words_pyvi)}")
print(f"Các từ: {words_pyvi[:20]}...")
```

## So sánh và thống kê kết quả

```python
import pandas as pd

# Đếm từ đơn và từ ghép trong kết quả pyvi
pyvi_single = [w for w in words_pyvi if '_' not in w]  
pyvi_compound = [w for w in words_pyvi if '_' in w]

# Bảng so sánh
data = {
    'Phương pháp': ['Naive', 'Từ điển lọc', 'pyvi'],
    'Từ đơn': [len(words_single), len(valid_single), len(pyvi_single)],
    'Từ ghép': [len(words_bigram), len(valid_bigram), len(pyvi_compound)],
    'Tổng': [len(words_single) + len(words_bigram), len(valid_single) + len(valid_bigram), len(words_pyvi)]
}

df = pd.DataFrame(data)
print(df.to_string(index=False))

print(f"\nKẾT LUẬN:")
print(f"- Naive: {len(words_single)} từ đơn, {len(words_bigram)} bigram (ghép tuần tự)")
print(f"- Từ điển: {len(valid_single)} từ đơn hợp lệ, {len(valid_bigram)} bigram có nghĩa")
print(f"- pyvi: {len(pyvi_single)} từ đơn, {len(pyvi_compound)} từ ghép (tách thông minh)")
```
